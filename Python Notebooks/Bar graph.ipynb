```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv('rainfall in india 1901-2015.csv', delimiter=',')
```

```python
x = data['SUBDIVISION']
y = data['ANNUAL']
```

```python

fig = plt.figure(figsize=(28, 10))
plt.bar(x,y, width = 0.8)
plt.xticks(rotation = 90)
plt.xlabel('States')
plt.ylabel('Annual rainfall 1901 - 2015 ')
#plt.legend()
plt.show()
plt.savefig("bar_graph.png")
```

```python
fig = plt.figure(figsize=(28, 10))
x = data['YEAR']
y = data['ANNUAL']
plt.plot(x,y)
plt.xlabel('ANI')
plt.ylabel('Rainfall 1901 - 2015')
plt.xticks(rotation = 90)
plt.show()
```
